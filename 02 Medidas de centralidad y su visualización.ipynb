```
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```
G = nx.gnp_random_graph(20, .3)
nx.draw(G)
```

```
etiquetas = dict(G.degree)
```

```
nx.draw(G,
        node_color = 'r',
        node_shape = 'o',
        node_size = 600,
        labels = etiquetas,
        with_labels = True)
```

```
tamano = np.array( list( nx.betweenness_centrality( G ).values( ) ))
tamano
```

```
nx.draw(G,
        node_color = 'y',
        node_shape = 'o',
        node_size = 10000*tamano,
        labels = etiquetas,
        with_labels = True)
```

```
colores = np.array(list(nx.closeness_centrality(G).values()))
colores
```

```

nx.draw(G,
        node_color = colores,
        node_shape = 'o',
        node_size = 10000*tamano,
        labels = etiquetas,
        with_labels = True, 
        cmap = 'hot',
        linewidths = 10)
```
